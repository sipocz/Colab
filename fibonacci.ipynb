```
def fibonacci(n):
    if n==0:
        return([])
    if n==1:
        return([0])
    if n==2:
        return([0,1])
    a=fibonacci(n-1)
    a.append(a[-1]+a[-2])
    return(a)    
```

```
print(fibonacci(3))
```
