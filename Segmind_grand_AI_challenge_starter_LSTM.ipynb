```python

from tensorflow.keras.layers import InputLayer, Dense, LSTM, Input, Dropout
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.optimizers import SGD,Adam,Adamax,Nadam,Ftrl,Adadelta
import tensorflow as tf
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.backend import clear_session
from tensorflow.keras.losses import mean_absolute_percentage_error
```

```python
XXX=df[df["Store"]==3 ][:]
```

```python
XXX.tail()
```

```python
XXX["Date_index"]=XXX["Timestamp"]
XXX.index=XXX["Date_index"]
```

```python
XXX.sort_index(inplace=True)
```

```python
XXX.reindex()
```

```python
prediktorok=["Promo","SchoolHoliday","a","b","c","zero","Day1","Day2","Day3","Day4","Day5","Day6","Day7","korr"]
weeklist=[]
for years in [2013,2014,2015]:
    for weeks in range(1,53):
        #print(years,weeks)
        a=XXX[(XXX["week_of_year"]==weeks) & (XXX["year"]==years)][prediktorok]
        if len(a)==6:
            weeklist.append(a)
            
            
```

```python
weeklist[0]
```

```python
'''
for elem in weeklist:
    if len(elem)==6:
        print(elem)
'''
NN_X=[]
NN_yy=[]

for aweek in weeklist: 
    #print((aweek[prediktorok]).values.tolist())
    #print(len(aweek))
    aaa=((aweek[prediktorok]).values.tolist())
    NN_X.append(aaa)
    suma=0
    for i in range(6):
        suma+=aaa[i][-1]
    
    NN_yy.append(suma)
```

```python
max_NNy=max(NN_y)
```

```python
NN_y=[a/max_NNy for a in NN_y]
```

```python
clear_session()



input_size=13
drop_frac0=0.7  
drop_frac1=0.7  

input1=Input(shape=(6,14))
lstm1=LSTM(116,dropout=0.1,return_sequences=False,)(input1)
l1_out=Dense(9256,activation="sigmoid",)(lstm1) # 
l2_out=Dropout(drop_frac0)(l1_out)


l3_out=Dense(9148,activation="sigmoid",)(l2_out)
l4_out=Dropout(drop_frac1)(l3_out)

pred=Dense(1,activation="sigmoid")(l4_out)

model = Model(inputs=input1, outputs=pred)
optimizer=Adamax(learning_rate=0.000001,) #

model.compile(loss='mae',
    optimizer=optimizer,
    metrics=["mae"])
```

```python
type(NN_X[1][1][0])
```

```python
 history = model.fit(NN_X[:-1], NN_y[1:], epochs=200, batch_size=32,
                    
                   verbose=1)

```

```python
datag=pd.DataFrame()
y_pred=model.predict(NN_X[:-1])

```
