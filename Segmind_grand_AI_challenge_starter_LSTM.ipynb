```python
clear_session()



input_size=13
drop_frac0=0.3  
drop_frac1=0.3  

input1=Input(shape=(input_size,))
#lstm1=LSTM(1,dropout=0.91,return_sequences=False)(input1)
l1_out=Dense(9600,activation="swish",)(input1) # 
l2_out=Dropout(drop_frac0)(l1_out)


l3_out=Dense(9000,activation="swish",)(l2_out)
l4_out=Dropout(drop_frac1)(l3_out)

pred=Dense(1,activation="swish")(l4_out)

model = Model(inputs=input1, outputs=pred)
optimizer=SGD(learning_rate=0.0005,) #

model.compile(loss='mse',
    optimizer=optimizer,
    metrics=["mse"])
```

```python
 history = model.fit(X_NN, y_NN, epochs=25, batch_size=3, validation_split = 0.2,
                    
                   verbose=1)

```
