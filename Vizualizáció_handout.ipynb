# Plot

MATLAB-jellegű, publikálható szinvonalú ábrák készítése. Elsősorban 2D, 3D statikus ábrák, de van animáció is. Kétféle használat: implicit, illetve egy MATLAB API jellegű objektum-orientált mód. Ez utóbbi bonyolultabb, de nagyobb szabadságot ad paraméter állításban. Ha több alábra kell, akkor ez a mód javasolt. 


```
# Matplotlib, plot
# Koordináta párok kirajzolása.

# Készítsük el az x koordinátához tartozó adatokat!
# Importáljuk be a numpy arange metódusát!
import numpy as np
# Hozzunk létre egy x vektors az np.arange segítségével 0 és 5 között, 0.2 lépésközzel!
# Dokumentáció: https://numpy.org/devdocs/reference/generated/numpy.arange.html
x=np.arange(0,5,0.2)
# Irassuk ki x-et!
x
```

```
# A matplotlib-bel fogunk dolgozni.
# Telepítsük fel matplotlib-et a !pip segítségével, ha nem lenne a gépen!
!ls
```

```
# Mivel a plot két koordinátával dolgozik, ezért szükségünk
# lesz egy y-ra is. Első körben adjuk értékül x-et y-nak!
y=x
# Importáljuk be a matplotlib.pyplot csomagot plt néven!
import matplotlib.pyplot as plt
# Plot-oljuk ki a plt.plot függvény használatával!
# Dokumentáció: https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot(x,y)
# Jelenítsük meg a plotot a plt.show függvény segítségével!
# Dokumentáció: https://matplotlib.org/3.3.2/api/_as_gen/matplotlib.pyplot.show.html
plt.show()
```

```
# Nézzük meg, hogy hogyan tudunk több adatot egy ábrára kirajzolni
# Adjuk y2-nek értékül x**2-t!
y2=x**2
# Adjuk y3-nek értékül x**3-t!
y3=x**3
# Plot-oljuk ki x és y párokat a plt.plot függvény használatával!
plt.plot(x,y)
# Plot-oljuk ki x és y2 párokat a plt.plot függvény használatával!
plt.plot(x,y2)
# Plot-oljuk ki x és y3 párokat a plt.plot függvény használatával!
plt.plot(x,y3,"g+")
# Jelenítsük meg a plotot a plt.show függvény segítségével!
plt.show()
```

```
# Vonal stílusok, legend
# Plot-oljuk ki x és y párokat 'r--' stílussal és 'lineáris' labellel!
plt.plot(x,y,"r--",label="linearis")
# Plot-oljuk ki x és y2 párokat 'bs' stílussal és 'négyzetes' labellel!
plt.plot(x,y2,"bs", label="négyzetes")
# Plot-oljuk ki x és y3 párokat 'g^' stílussal és 'köbös' labellel!
plt.plot(x,y3,"g^", label="köbös")
# Állítsuk be a legendet a bal felső sarokba a plt.legend segítségével!
# Dokumentáció: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html

plt.legend(loc="upper center") # upper left, center, right
# Jelenítsük meg a plotot a plt.show függvény segítségével!
plt.show()
```

```
# Seaborn alap színkonfigurációi is szépek, jól használhatók

# Plot color palette metódus definíciója
def plot_color_palette(palette):   
    # Dokumentáció: https://www.kite.com/python/docs/seaborn.palplot
    figure = sns.palplot(sns.color_palette())
    plt.xlabel("Color palette: " + palette)
    plt.show(figure)

# Paletták kirajzolása
palettes = ["deep", "muted", "pastel", "bright", "dark", "colorblind"]
for palette in palettes:
    sns.set(palette=palette)
    plot_color_palette(palette)
```

# Adatok generálása

Gyakran dolgozunk eloszlásokkal. Az eloszlások segítenek például a zajszűrésben, illetve bizonyos esetekben jobban sikerül a modellillesztés, ha nem egzakt értékekkel, hanem eloszlásokkal dolgozunk.

Alapvető funkcionalitás adott eloszlású adatok generálása. Inkább elméleti céllal használjuk, illetve, ha nincs adatunk. De amióta van a Kaggle, azóta adat is van.

Vizualizáció célja gyakran egyváltozós és két változós eloszlások ábrázolása (hisztogram és együttes eloszlás, illetve annak marginálisai).

```
# Adatok generálása adott eloszlásból.
# Paraméterek: mean, covariancia mátrix (szimmetrikus, pozitív definit), mintaszám
data = np.random.multivariate_normal([0, 1], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['a', 'b'])
data
```

```
# Két eloszlásból származó adatok megjelenítése.
# Így könnyű, mivel két külön oszlopban vannak. :)
for col in 'ab':
    plt.hist(data[col], alpha=0.5, density=True)
```

```
# Kernel density estimation ~ az adatok simítása
# https://en.wikipedia.org/wiki/Kernel_density_estimation
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```
