```
def nums(n):
    i=1
    j=1
    num=1
    while num < n:
        
        if num==1 or num==2:
            yield  1
            i=1
            j=num
            k=i+j
        else: 
            yield k
            k=i+j   
        num+=1
        (i,j)=(j,k)    
    
```

```
nums(2)
```

```
for i in nums(3000):
    print(i)
```
