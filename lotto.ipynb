```
openfile=open('/content/drive/My Drive/lotto/otos.csv')
numtable=[]
for line in openfile:
    #print(line)
    line=line.strip()
    picdata=line.split(";")
    rec=[int(picdata[-5]),int(picdata[-4]),int(picdata[-3]),int(picdata[-2]),int(picdata[-1])]
    numtable.append(rec)
    #print(i," ")
openfile.close()
```

```
table={}
for i in range(1,91):
    table[i]=0
```

```
for huz in numtable:
    for num in huz:
        table[num]=table[num]+1
        
```

```
numtable.reverse()
print(numtable[0]) #16,64,71,77,89
```

```
flist=[]
for i in range(len(numtable)-1):
    #print (numtable[i],"-->", numtable[i+1])
    for a in numtable[i]:
        for b in numtable[i+1]:
            flist.append((a,b))
```

```
statistic={}
for i in range(1,91):
    for j in range(1,91):
        statistic[(i,j)]=0

```

```
for i in flist:
    statistic[i]+=1
    
```

```
for i in statistic:
    print(i,":",statistic[i],end="__",sep="")
```

```
maxi=max(statistic.values())
for imin in range(maxi,maxi-4,-1):
    for i in statistic:
        if statistic[i]==imin:
            print(i,imin)
        

```

```
current=[20,36,54,72,76]
bestlist={}

for i in current:
    for j in range(1,91): 
        s=(i,j)
        bestlist[s]=statistic[s]
print(bestlist)
max=0
for i in bestlist:
    if bestlist[i]>max:
        print (i, bestlist[i])
        max=bestlist[i]
```
