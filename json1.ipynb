```
import urllib
from urllib import request
res=request.urlopen("https://isic-archive.com/api/v1/image?limit=60000&offset=0&sort=name&sortdir=1&detail=false")
print (res)
```

```
res.code
t=res.read()
```

```
print(t)
```

```
t=t.decode("utf-8")
print(t)
```

```
import json
data=json.loads(t)
print(data)
```

```
tb=[]
for i in data:
    #print(i["_id"], i[name])
    tb.append([i["_id"], i["name"]])
print(tb)
```
