```
class Line:
    
    def __init__(self,coor1,coor2):
        self.a=coor1
        self.b=coor2
        self.dist=((self.a[0]-self.b[0])**2+(self.a[1]-self.b[1])**2)**0.5
        self.slop=(self.b[1]-self.a[1])/(self.b[0]-self.a[0])
        
    def distance(self):
        return(self.dist)
        
    def slope(self):
        return(self.slop)
```

```
class Cylinder:
    pi=3.1415926
    def __init__(self,height=1,radius=1):
        self.height=height
        self.radius=radius
        
        
    def volume(self):
        self.vol=(2*self.radius)**2*self.pi/4*self.height
        return(self.vol)
        pass
    
    def surface_area(self):
        self.surf=2*self.radius*self.pi*self.height+(2*self.radius)**2*self.pi/4*2
        return(self.surf)
    
```
