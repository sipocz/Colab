```
def splitOUTfile(fname):
    '''
    :param fname: IN file neve, ehhez keresi a MALDI OUT file-t
    :return:  az illeszkedő fájlok listája
    '''
    fname_list=fname.split(".")
    fname_1=fname_list[0]
    parts=fname_1.split("-")
    return([parts[2],fname])

```

```
def splitOutfiles(fnamelist):
    out=[]
    for fname in fnamelist:
        out.append(splitOUTfile(fname))
    return(out)
```
