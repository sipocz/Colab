## Adat sorrend függőség

A klaszterek általában egy iteratív algoritmus segítségével állnak elő. Ritka az explicit megoldás. Iterálni általában a hibafelületen szoktunk, azonban az is kérdés, hogy hogyan számítjuk ki az aktuális hibát (a modell paramétereinek függvényében).

A hiba kiszámítása történhet úgy is, hogy vesszük a hiba átlagát az összes pont viszonylatában. Ekkor a tanítás független az adatok sorrendjétől. Ez gyakorlatilag az a módszer, aminek a legmagasabb a számításigénye. Ha fel szeretnénk gyorsítani az algoritmust, alkalmazhatunk sampling technikát, azaz a tanítóadatok egy részhalmazát vesszük. A sampling technika így befolyásolja a végeredményt, azaz, hogy milyen klasztereket alakítunk ki.

Az algoritmusokat két főbb csoportba osztjuk: top-down és bottom-up. A top-down módszerek a teljes mintából indulnak, majd elkezdik feldarabolni. A bottom-up módszerek pontból / pontokból indulnak, és elkezdik összevonni. Az, hogy milyen pontokkal kezdünk, befolyásolja a végső klasztereket.

Somma: bizonyos esetekben a végső klaszterek függhetnek az adatok sorrendjétől. Ha van idő, érdemes lehet egy másik permutációt is megnézni.

## Hard (exkluzív) vagy "soft", fuzzy csoportok

Lehetséges, hogy nem csak arra van szükségünk, hogy egy pont melyik klaszterbe tartozik (, és melyikbe nem), hanem arra, hogy mennyire tartozik egy klaszterbe. Ilyenkor a klaszter tagság nem egy bináris döntés ($\{0,1\}$), hanem egy 0 és 1 közötti szám ($[0,1]$).

## A priori tudás feltevése, felhasználó által definiált paraméterek

Általában sietünk, tehát általában szeretjük, ha nem kell paramétert megadni a klaszterező algoritmus számára. Ilyen paraméterek például: hány klaszterem van, meddig iteráljon az algoritmus, mi legyen a maximális távolság két pont között, milyen eloszlást használjon? A leginkább azt szeretnénk, ha az algoritmus kitalálná például, hogy hány klaszterre oszthatók a pontjaim. Van ilyen algoritmus is.

Nézhetjük a másik oldalról is: minél több priori tudást / információt viszünk be a rendszerbe, annál jobb minőséget tudunk elvárni az algoritmustól. Megpróbálhatunk segíteni, hogy minél jobban végezze el a feladatát.

Az adott helyzettől függ, hogy éppen mire van igény. Érdemes a fentieket mérlegelni egy adott feladat esetén.

## Konvergencia

A k-means algoritmus egy relatíve egyszerűbb módszer. A lényege, hogy megkeresi a megfelelő centroidokat, amelyek illeszkednek az adatokra. A klaszterek gyakorlatilag a centroidok adott környezetében levő pontok. A k-means algoritmus konvergenciáját többen is vizsgálták. Az optimum keresés az EM eljárással történik. ## Ábrák

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.figure(figsize=(12, 12))

n_samples = 1500
random_state = 170   
X, y = make_blobs(n_samples=n_samples, random_state=random_state)

# Incorrect number of clusters
y_pred = KMeans(n_clusters=2, random_state=random_state).fit_predict(X)

plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis')
plt.title(u"Helytelen cluster szám")

# Anisotropicly distributed data
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]
X_aniso = np.dot(X, transformation)
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_aniso)

plt.subplot(222)
plt.scatter(X_aniso[:, 0], X_aniso[:, 1], c=y_pred,cmap='viridis')
plt.title(u"Anizotróp eloszlás")

# Different variance
X_varied, y_varied = make_blobs(n_samples=n_samples,
                                cluster_std=[1.0, 2.5, 0.5],
                                random_state=random_state)
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_varied)

plt.subplot(223)
plt.scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred,cmap='viridis')
plt.title(u"Eltérő szórás")

# Unevenly sized blobs
X_filtered = np.vstack((X[y == 0][:500], X[y == 1][:100], X[y == 2][:10]))
y_pred = KMeans(n_clusters=3,
                random_state=random_state).fit_predict(X_filtered)

plt.subplot(224)
plt.scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred,cmap='viridis')
plt.title(u"Eltérő cluster méret")

plt.show()
```

# Távolság, hasonlóság mérése

* Matek, a távolságra igaznak kell lennie a következőknek:
 * $0\leq d$
 * $d(a,b)=d(b,a)$
 * $d(a,b)+d(b,c) \geq d(a,c)$ 
* Minkowski-távolság, $L^p$-norma általános definíciója:
 * $X=(x_1,x_2,...,x_n)\in R^n$ és $Y=(y_1,y_2,...,y_n)\in R^n$ 
 * $d(X,Y)_p=(\Sigma_{i=1}^n |x_i-y_i|^p)^{1/p}$
 * $p=2\Longrightarrow$  Euclides-i távolság, L2
 * $p=1\Longrightarrow$  Manhattan távolság, L1
 * $p=\infty\Longrightarrow$  Lmax, Csebisev-távolság
 * Egységsugarú kört lásd lent, különböző p értékek esetén.

![Vektor_normák](https://drive.google.com/uc?export=view&id=14Zuy78ZFdNmn4988_Fi5SI3ZYGLgSrSP)

* Egyéb távolság metrikák:
 * Pearson korreláció
 * Spearman rang korreláció - lásd: https://en.wikipedia.org/wiki/Spearman%27s_rank_correlation_coefficient
 * Távolság korreláció - lásd: https://en.wikipedia.org/wiki/Distance_correlation
 * Koszinusz hasonlóság: $cos(\theta) =\frac{XY}{\| X \|_2 \| Y \|_2}$ - két (nem null) vektor szögének koszinusza. (nem "igazi távolság", háromszög egyenlőtlenség nem teljesül)

## "Lapos" módszerek

### K-means

* Gyakran használt, elterjedt módszer.
* Klaszter középpontokkal (mean / centroid) dolgozik, a modell gyakorlatilag a középpontok koordinátáit jelenti.
* Iteratív módszer
 0. Kezdetben véletlenszerűen vannak beállítva a klaszter középpontok.
 1. Minden mintapontot rendeljünk hozzá egy klaszter középponthoz a távolságuk alapján. Minden mintapont a legközelebbi középponthoz kerül.
 2. Számoljuk újra a középpontokat a klaszterhez hozzárendelt mintapontok alapján.
 3. Goto 1.
 4. Az algoritmus akkor terminál, ha már nem változnak a klaszterhez való hozzárendelések, azaz beálltak a középpontok.
* Variánsok
 * K-medians: kevésbé zaj/outlier függő. A median függvény használata általában ezzel jár.
 * Fuzzy C-means. "Soft" módszer, minden adatpont több clusterhez is tartozhat.

### Hierarchikus fajtái

* Agglomerative (összevonó) módszerek: AGNES, CURE, CHAMELEON
 * Algoritmus példa
   * Elöször minden pont külön klaszter.
   * Elkezdem iteratívan összevonni a pontokat.
   * Mindig abba az irányba megyek, ahol a klaszter távolság (lásd fent) javul.
* Divisive (felosztó) módszerek: DIANA
 * Algoritmus példa
   * Elöször egy klaszterem van, amibe az összes pont beletartozik.
   * Elkezdem felosztani a pontokat.
   * Mindig abba az irányba megyek, ahol a klaszter távolsága legjobb.

## DBSCAN

Density-based spatial clustering of applications with noise (1996, Test of time award 2014)

Definíció

* Core point: Az a minta elem amelynek az adott sugarú környezetében legalább adott darab minta van.
* Közvetlen elérhetőség (Directly reachable): core point környezetében van
* Elérhetőség (Reachable): tranzitivitás a közvetlen elérhetőség reláción keresztül
* Egy cluster, ami egy core-ból közvetlenül elérhető

Tulajdonságok
    
* $+$ Cluster számot nem kell előre megadni
* $+$ Tetszőleges alakú clustereket is megtalál 
* $+$ Nem érzékeny outlierre
* $+$ Jól skálázható,
* $-$ Nem teljesen determinisztikus (széleken lévő pontok különböző clusterekbe eshetnek)
* $-$ Nagyon különböző sűrűségű problémákra nem működik rögzített paraméterekkel
* $-$ Jóság függ hasonlósági metrikától

# Nagy-dimenziós módszerek

## Subspace clustering

CLIQUE

A módszer paraméterei:
* xi - Hány egyenlő részre vágja a teret
* tau - Mekkora a minimális sűrűség.

Elöször dimenziónként vágja fel a teret. Ha az egyes szeletekben elegendő (tau)mennyiségű mintapont van, akkor megtartjuk. Ha ez megvolt, akkor megnézzük a 2 dimenziós esetet, azaz a dimenzió párokat. Itt is azokat tartjuk meg, ahol elegendő adat van. Majd folytatjuk 3, 4, ... dimenzióban.

![](https://upload.wikimedia.org/wikipedia/commons/8/86/SubspaceClustering.png)

## "Hubness" based clustering

A hubness azt a jelenséget írja le, hogy a magas dimenziós adatokban is megfigyelhetők bizonyos tendenciák, mint például az, hogy az adatpontok néhány helyen, bizonyos adatpontok környékén sűrűsödnek. Ezeket az adatopontokat Hub-nak nevezzük.

Az ábra három kiemelt (virtuális) pontot illusztrál:
* C - centroid - a "tömegközéppont"
* M - medoid - a "magas dimenziós medián"
* H - hub - a pont, aki körül sokan vannak

# Scikit-learn eszköztára

```python
import time
import warnings

import numpy as np
import matplotlib.pyplot as plt

from sklearn import cluster, datasets, mixture
from sklearn.neighbors import kneighbors_graph
from sklearn.preprocessing import StandardScaler
from itertools import cycle, islice

np.random.seed(0)

# ============
# Generate datasets. We choose the size big enough to see the scalability
# of the algorithms, but not too big to avoid too long running times
# ============
n_samples = 1500
noisy_circles = datasets.make_circles(n_samples=n_samples, factor=.5,
                                      noise=.05)
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=.05)
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)
no_structure = np.random.rand(n_samples, 2), None

# Anisotropicly distributed data
random_state = 170
X, y = datasets.make_blobs(n_samples=n_samples, random_state=random_state)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)
aniso = (X_aniso, y)

# blobs with varied variances
varied = datasets.make_blobs(n_samples=n_samples,
                             cluster_std=[1.0, 2.5, 0.5],
                             random_state=random_state)

# ============
# Set up cluster parameters
# ============
plt.figure(figsize=(9 * 2 + 3, 12.5))
plt.subplots_adjust(left=.02, right=.98, bottom=.001, top=.96, wspace=.05,
                    hspace=.01)

plot_num = 1

default_base = {'quantile': .3,
                'eps': .3,
                'damping': .9,
                'preference': -200,
                'n_neighbors': 10,
                'n_clusters': 3}

datasets = [
    (noisy_circles, {'damping': .77, 'preference': -240,
                     'quantile': .2, 'n_clusters': 2}),
    (noisy_moons, {'damping': .75, 'preference': -220, 'n_clusters': 2}),
    (varied, {'eps': .18, 'n_neighbors': 2}),
    (aniso, {'eps': .15, 'n_neighbors': 2}),
    (blobs, {}),
    (no_structure, {})]

for i_dataset, (dataset, algo_params) in enumerate(datasets):
    # update parameters with dataset-specific values
    params = default_base.copy()
    params.update(algo_params)

    X, y = dataset

    # normalize dataset for easier parameter selection
    X = StandardScaler().fit_transform(X)

    # estimate bandwidth for mean shift
    bandwidth = cluster.estimate_bandwidth(X, quantile=params['quantile'])

    # connectivity matrix for structured Ward
    connectivity = kneighbors_graph(
        X, n_neighbors=params['n_neighbors'], include_self=False)
    # make connectivity symmetric
    connectivity = 0.5 * (connectivity + connectivity.T)

    # ============
    # Create cluster objects
    # ============
    ms = cluster.MeanShift(bandwidth=bandwidth, bin_seeding=True)
    two_means = cluster.MiniBatchKMeans(n_clusters=params['n_clusters'])
    ward = cluster.AgglomerativeClustering(
        n_clusters=params['n_clusters'], linkage='ward',
        connectivity=connectivity)
    spectral = cluster.SpectralClustering(
        n_clusters=params['n_clusters'], eigen_solver='arpack',
        affinity="nearest_neighbors")
    dbscan = cluster.DBSCAN(eps=params['eps'])
    affinity_propagation = cluster.AffinityPropagation(
        damping=params['damping'], preference=params['preference'])
    average_linkage = cluster.AgglomerativeClustering(
        linkage="average", affinity="cityblock",
        n_clusters=params['n_clusters'], connectivity=connectivity)
    birch = cluster.Birch(n_clusters=params['n_clusters'])
    gmm = mixture.GaussianMixture(
        n_components=params['n_clusters'], covariance_type='full')

    clustering_algorithms = (
        ('MiniBatchKMeans', two_means),
        ('AffinityPropagation', affinity_propagation),
        ('MeanShift', ms),
        ('SpectralClustering', spectral),
        ('Ward', ward),
        ('AgglomerativeClustering', average_linkage),
        ('DBSCAN', dbscan),
        ('Birch', birch),
        ('GaussianMixture', gmm)
    )

    for name, algorithm in clustering_algorithms:
        t0 = time.time()

        # catch warnings related to kneighbors_graph
        with warnings.catch_warnings():
            warnings.filterwarnings(
                "ignore",
                message="the number of connected components of the " +
                "connectivity matrix is [0-9]{1,2}" +
                " > 1. Completing it to avoid stopping the tree early.",
                category=UserWarning)
            warnings.filterwarnings(
                "ignore",
                message="Graph is not fully connected, spectral embedding" +
                " may not work as expected.",
                category=UserWarning)
            algorithm.fit(X)

        t1 = time.time()
        if hasattr(algorithm, 'labels_'):
            y_pred = algorithm.labels_.astype(np.int)
        else:
            y_pred = algorithm.predict(X)

        plt.subplot(len(datasets), len(clustering_algorithms), plot_num)
        if i_dataset == 0:
            plt.title(name, size=18)

        colors = np.array(list(islice(cycle(['#377eb8', '#ff7f00', '#4daf4a',
                                             '#f781bf', '#a65628', '#984ea3',
                                             '#999999', '#e41a1c', '#dede00']),
                                      int(max(y_pred) + 1))))
        # add black color for outliers (if any)
        colors = np.append(colors, ["#000000"])
        plt.scatter(X[:, 0], X[:, 1], s=10, color=colors[y_pred])

        plt.xlim(-2.5, 2.5)
        plt.ylim(-2.5, 2.5)
        plt.xticks(())
        plt.yticks(())
        plt.text(.99, .01, ('%.2fs' % (t1 - t0)).lstrip('0'),
                 transform=plt.gca().transAxes, size=15,
                 horizontalalignment='right')
        plot_num += 1

plt.show()
```

# Kiértékelés

## Belső metrikák

Cél: magas intra-cluster hasonlóság (alacsony scatter, $S_i$), alacsony inter-cluster hasonlóság (separation, $M_{ij}$), illetve stabilitás (pár adatpont törlése vagy perturbációja ne okozzon nagy változást)

Scatter és separation definíció: https://sigmodrecord.org/publications/sigmodRecord/0209/a1.partii_clvalidity1.pdf

Példák metrikákra:
* Davies–Bouldin index: $f(\frac{S_i+S_j}{M_{ij}})$
* Dunn index $f(\frac{min(M_{ij})}{max(S_i)})$
* Silhouette Coefficient $s(i) = \frac{b(i)-a(i)}{max(a(i),b(i))}$, ahol $a(i)$ átlagos eltérés saját cluster elemeitől, $b(i)$ legalacsonybb átlagos eltérés más clusterektől. Mivel ez minden elemre ad jellemzést, ezért grafikus képet ad clusterek hasonlóságáról
* SSE: Sum of Squared Error:  minták clusterközéptől vett eltérés négyzetének összege
* Calinski-Harabaz Index - https://scikit-learn.org/stable/modules/generated/sklearn.metrics.calinski_harabasz_score.html

## Külső metrikák

Van Ground Truth címke az adatpontok egy részéhez

* Purity: az egy clusterben levő mintapontok egy osztályba tartoznak-e? Nem bünteti, ha túl sok cluster van.
* Fowlkes-Mallows index: osztályozási analógia, precision és recall mértani közepe [osztályozáshoz tartozó kiértékelésről később beszélünk részletesen]
* Adjusted Rand index: osztályozási analógia, bináris keveredési mátrix alapján (véletlen találattal korrigálva)
* Adjusted Mutual Information

# Klaszterszám meghatározása

## Elbow method

Grid-search, "elbow method": különböző cluster szám feltevés mellett clusterezzük az adatpontokat és kiszámoljuk valamelyik index értéket a fentiek közül

Alternatíva: Cross-validation (kxCV) alapú stabilitás vizsgálat: teszt adatpontok átlagos távolsága a legközelebbi clustertől. https://arxiv.org/pdf/1702.02658.pdf

```python
# clustering dataset
# determine k using elbow method
# https://pythonprogramminglanguage.com/kmeans-elbow-method/ kicsit módosítva
 
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
 
x1 = np.array([3, 1, 1, 2, 1, 6, 6, 6, 5, 6, 7, 8, 9, 8, 9, 9, 8])
x2 = np.array([5, 4, 5, 6, 5, 8, 6, 7, 6, 7, 1, 2, 1, 2, 3, 2, 3])
s = 25

c=['b','b','b','b','b','g','g','g','g','g','r','r','r','r','r','r','r']

 
plt.plot()
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.title('Dataset')
plt.scatter(x1, x2,s,c)
plt.show()

 
# create new plot and data
plt.plot()
X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)

 
# k means determine k
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    distortions.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X.shape[0])

# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('Az Elbow módszer kiadja a helyes k értéket')
plt.show()
```

## Sziluett analízis

* A sziluett analízis segítségével a klaszterek közti szeparáció mértékét tanulmányozhatjuk.
* Minden mintapont esetén megmutatja, hogy milyen közel van a klaszter többi pontjához.
* Leolvasható a klaszterszám, illetve látható a távolságok tapasztalati eloszlása.

Részletek: https://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html

```python
from __future__ import print_function

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

print(__doc__)

# Generating the sample data from make_blobs
# This particular setting has one distinct cluster and 3 clusters placed close
# together.
X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=4,
                  cluster_std=1,
                  center_box=(-10.0, 10.0),
                  shuffle=True,
                  random_state=1)  # For reproducibility

range_n_clusters = [2, 3, 4, 5, 6]

for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1 but in this example all
    # lie within [-0.1, 1]
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    # This gives a perspective into the density and separation of the formed
    # clusters
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.Spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.Spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(X[:, 0], X[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c=colors, edgecolor='k')

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st feature")
    ax2.set_ylabel("Feature space for the 2nd feature")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

    plt.show()
```
