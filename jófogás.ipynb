```
import requests
from bs4 import  BeautifulSoup


def querycity(city):
    print("------------------------------------------------")
    print("+    ",city.capitalize())


 
    url="https://ingatlan.jofogas.hu/"+city+"/haz?st=s"
    res=requests.get(url)
    #print (res.text)
    soup=BeautifulSoup(res.text,"html.parser")
  
    urllist=soup.find_all("a", class_="subject")
    #print(urllist)
    pricelist=soup.find_all("span",itemprop="price")
    #print(pricelist)


    out1=[]
    out2={}
    for i in enumerate(urllist[1:]):
        link=""+i[1]['href']
        print(f"+{i[0]:>4}. {pricelist[i[0]].text:>12} Ft :  {link:<1000}")
        #print(i[0],". ",pricelist[i[0]].text,link)
        out1.append([pricelist[i[0]].text,link])

        #out.append([urllist[i],pricelist[i]])
    
    print("----------------  END  ------------------------")


querycity("komarom-esztergom/esztergom")
querycity("komarom-esztergom/mogyorosbanya")
querycity("pest/piliscsaba")
querycity("komarom-esztergom/csolnok")
querycity("pest/zsambek")
querycity("komarom-esztergom/gyermely")
querycity("komarom-esztergom/szomor")
querycity("pest/csobanka")
querycity("pest/tinnye")
querycity("pest/pilisvorosvar")
querycity("pest/solymar")
querycity("pest/perbal")
querycity("pest/telki")









```
